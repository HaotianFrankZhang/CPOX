```python
def calcDistances262_408(numbers, currDirectory):
    distances_A = []
    distances_B = []
    for currNumber in numbers:
        structure = parsePDB(currDirectory + 'frame_' + currNumber + '.pdb')
        # Select the alpha carbon atoms of the specified residues
        ca1 = structure.select('resid 408 and name CA')
        ca2 = structure.select('resid 262 and name CA')

        # Ensure the selections are not empty
        distance_A = calcDistance(ca1[0], ca2[0])
        distance_B = calcDistance(ca1[1], ca2[1])
        distances_A.append(distance_A)
        distances_B.append(distance_B)
    return distances_A, distances_B
```

```python
def calcDistances340_409(numbers, currDirectory):
    distances_A = []
    distances_B = []
    for currNumber in numbers:
        structure = parsePDB(currDirectory + 'frame_' + currNumber + '.pdb')
        # Select the alpha carbon atoms of the specified residues
        ca1 = structure.select('resid 340 and name CA')
        ca2 = structure.select('resid 409 and name CA')

        # Ensure the selections are not empty
        distance_A = calcDistance(ca1[0], ca2[0])
        distance_B = calcDistance(ca1[1], ca2[1])
        distances_A.append(distance_A)
        distances_B.append(distance_B)
    return distances_A, distances_B
```

```python
def parse_features(file_path):
    features = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split('\t')
            if parts[0] != 'frame':
                frame_pocket = (int(parts[0].split('_')[1]), parts[1].strip())  # Tuple of (frame, pocket)
                # Extracting features as a dict
                feature_values = dict(part.split(': ') for part in parts[2:])
                features[frame_pocket] = feature_values
    return features

# Parse the file containing the pockets of interest
def parse_pockets_of_interest(file_path):
    pockets_of_interest = {}
    countList = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split('\t')
            frame = parts[0]
            pocket = parts[1]
            ratio = parts[2].split(',')
            if frame != 'frame':
                pockets_of_interest[int(frame.split('_')[1])] = pocket
                countList[int(frame.split('_')[1])] = [float(x) for x in ratio]
    return pockets_of_interest, countList

# Parse the file containing the pockets of interest
def parse_pockets_of_interest(file_path):
    pockets_of_interest = {}
    countList = {}
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split('\t')
            frame = parts[0]
            pocket = parts[1]
            ratio = parts[2].split(',')
            if frame != 'frame':
                pockets_of_interest[int(frame.split('_')[1])] = pocket
                countList[int(frame.split('_')[1])] = [float(x) for x in ratio]
    return pockets_of_interest, countList


def FeatureDistribution(features, pockets_of_interest, cutoff_count, cutoffs):
    frames = sorted(pockets_of_interest.keys(), key=lambda x: int(x))  # Sort frames by number
    
    volumes_cutoff = [float(features[(frame, pockets_count[frame])]['Volume']) \
                      for frame in frames if (frame, pockets_count[frame]) in features and \
                      (cutoff_count[frame][0]/cutoff_count[frame][1]) >= cutoffs]


    TotalSASA = [float(features[(frame, pockets_count[frame])]['Total SASA']) \
                      for frame in frames if (frame, pockets_count[frame]) in features and \
                      (cutoff_count[frame][0]) >= cutoffs]

    HydroDensity = [float(features[(frame, pockets_count[frame])]['Mean local hydrophobic density']) \
                      for frame in frames if (frame, pockets_count[frame]) in features and \
                      (cutoff_count[frame][0]) >= cutoffs]

    PolarSASA = [float(features[(frame, pockets_count[frame])]['Polar SASA']) \
                      for frame in frames if (frame, pockets_count[frame]) in features and \
                      (cutoff_count[frame][0]) >= cutoffs]
    # plotGram(PolarSASA, frames, 'Polar SASA')

    charge = [float(features[(frame, pockets_count[frame])]['Charge score']) \
                      for frame in frames if (frame, pockets_count[frame]) in features and \
                      (cutoff_count[frame][0]) >= cutoffs]
    
    return volumes_cutoff, TotalSASA, PolarSASA, HydroDensity, charge


def evalResults_distribution(chain1_features_ratio, features, pockets_ratio, cutoff_ratio, cutoffs):
    volumes_cutoff, TotalSASA, PolarSASA, HydroDensity, charge = \
    FeatureDistribution(features, pockets_ratio, cutoff_ratio, cutoffs)
    chain1_features_ratio[currMutation][run]['Volumes'] = volumes_cutoff
    chain1_features_ratio[currMutation][run]['TotalSASA'] = TotalSASA
    chain1_features_ratio[currMutation][run]['PolarSASA'] = PolarSASA
    chain1_features_ratio[currMutation][run]['Hydrophobic Density'] = HydroDensity
    chain1_features_ratio[currMutation][run]['Charge'] = charge
    # print (chain1_features_ratio)
    return chain1_features_ratio        
```

```python
def extractFeatures(features, pockets_of_interest, cutoff_count):
    frames = sorted(pockets_of_interest.keys(), key=lambda x: int(x))  # Sort frames by number
    volumes = [float(features[(frame, pockets_of_interest[frame])]['Volume']) for frame in frames if
                   (frame, pockets_of_interest[frame]) in features]

    volumes_normalize = [float(features[(frame, pockets_of_interest[frame])]['Volume']) \
                         * (cutoff_count[frame][0]/cutoff_count[frame][1]) \
                         for frame in frames if
                   (frame, pockets_of_interest[frame]) in features]


    TotalSASA = [float(features[(frame, pockets_of_interest[frame])]['Total SASA']) for frame in frames if
                   (frame, pockets_of_interest[frame]) in features]
    # plotGram(TotalSASA, frames, 'SASA')

    HydroDensity = [float(features[(frame, pockets_of_interest[frame])]['Mean local hydrophobic density']) for frame in frames if
                     (frame, pockets_of_interest[frame]) in features]

    PolarSASA = [float(features[(frame, pockets_of_interest[frame])]['Polar SASA']) for frame in frames if
                         (frame, pockets_of_interest[frame]) in features]
    # plotGram(PolarSASA, frames, 'Polar SASA')

    charge = [float(features[(frame, pockets_of_interest[frame])]['Charge score']) for frame in frames if
                             (frame, pockets_of_interest[frame]) in features]
    
    return volumes, volumes_normalize, TotalSASA, PolarSASA, HydroDensity, charge

def evalResults(chain1_features_ratio, features, pockets_ratio, cutoff_ratio):
    volumes, volumes_normalize, TotalSASA, PolarSASA, HydroDensity, charge = \
    extractFeatures(features, pockets_ratio, cutoff_ratio)
    chain1_features_ratio[currMutation][run]['Volumes'] = volumes
    chain1_features_ratio[currMutation][run]['Volumes_normalize'] = volumes_normalize
    chain1_features_ratio[currMutation][run]['TotalSASA'] = TotalSASA
    chain1_features_ratio[currMutation][run]['PolarSASA'] = PolarSASA
    chain1_features_ratio[currMutation][run]['HydroDensity'] = HydroDensity
    chain1_features_ratio[currMutation][run]['Charge'] = charge
    return chain1_features_ratio
```

```python
chain1_features_ratio = {}
chain2_features_ratio = {}
chain1_features_count = {}
chain2_features_count = {}

for currMutation in mutations:
    if currMutation not in chain1_features_ratio:
        chain1_features_ratio[currMutation] = {}
    if currMutation not in chain2_features_ratio:
        chain2_features_ratio[currMutation] = {}
    if currMutation not in chain1_features_count:
        chain1_features_count[currMutation] = {}
    if currMutation not in chain2_features_count:
        chain2_features_count[currMutation] = {}
    
    for run in Runs:
        if run not in chain1_features_ratio[currMutation]:
            chain1_features_ratio[currMutation][run] = {}
        if run not in chain2_features_ratio[currMutation]:
            chain2_features_ratio[currMutation][run] = {}
        if run not in chain1_features_count[currMutation]:
            chain1_features_count[currMutation][run] = {}
        if run not in chain2_features_count[currMutation]:
            chain2_features_count[currMutation][run] = {}
            
        features_file = './' + currMutation + '/' + run + '_Features.txt'
        pockets_of_interest_file = './' + currMutation + '/fpocketResults/' + run + '_chain1.txt'
        pockets_ratio_file = './' + currMutation + '/fpocketResults/Ratio_' + run + '_chain1.txt'

        # Parsing files
        features = parse_features(features_file) # {(frame, pocket): {scores: 'numbers'}}
        pockets_count, cutoff_count = parse_pockets_of_interest(pockets_of_interest_file) # {#frame: pocket}
        pockets_ratio, cutoff_ratio = parse_pockets_of_interest(pockets_ratio_file)

        chain1_features_ratio = \
        evalResults(chain1_features_ratio, features, pockets_ratio, cutoff_ratio)
        
        chain1_features_count = \
        evalResults(chain1_features_count, features, pockets_count, cutoff_count)

        pockets_of_interest_file = './' + currMutation + '/fpocketResults/' + run + '_chain2.txt'
        pockets_ratio_file = './' + currMutation + '/fpocketResults/Ratio_' + run + '_chain2.txt'

        # Parsing files
        pockets_count, cutoff_count = parse_pockets_of_interest(pockets_of_interest_file) # {#frame: pocket}
        pockets_ratio, cutoff_ratio = parse_pockets_of_interest(pockets_ratio_file)
        
        chain2_features_ratio = \
        evalResults(chain2_features_ratio, features, pockets_ratio, cutoff_ratio)
        
        chain2_features_count = \
        evalResults(chain2_features_count, features, pockets_count, cutoff_count)
```

```python
def find_top_indices(allDistance, allVolume):
    # Results dictionary
    results = {}

    # Iterate over each mutation in the allDistance dictionary
    for mutation in allDistance:
        results[mutation] = {}

        # Iterate over each run in this mutation
        for run in allDistance[mutation]:
            for chain in allDistance[mutation][run]:
                # Collect distances
                distances = allDistance[mutation][run][chain]
                # Collect corresponding volumes from allVolume
                volumes = allVolume[chain][mutation][run]['Volumes']

                # Find indices where distances are 9 or more
                valid_indices = [i for i, d in enumerate(distances) if d >= 10]

                # Get corresponding volumes for these indices
                corresponding_volumes = [volumes[i] for i in valid_indices]

                # Create tuples of (volume, index) for sorting
                volume_with_index = list(zip(corresponding_volumes, valid_indices))

                # Sort based on volume (largest first) and get the top 10
                top_indices_by_volume = sorted(volume_with_index, reverse=True, key=lambda x: x[0])[:10]

                # Save results
                results[mutation][f"{run}_{chain}"] = [idx for _, idx in top_indices_by_volume]

    return results
```

```python
def find_top_indices_with_values(allDistance, allVolume):
    # Results dictionary
    results = {}

    # Iterate over each mutation in the allDistance dictionary
    for mutation in allDistance:
        results[mutation] = {}

        # Iterate over each run in this mutation
        for run in allDistance[mutation]:
            for chain in allDistance[mutation][run]:
                # Collect distances
                distances = allDistance[mutation][run][chain]
                # Collect corresponding volumes from allVolume
                volumes = allVolume[chain][mutation][run]['Volumes']

                # Find indices where distances are 9 or more
                valid_indices = [i for i, d in enumerate(distances) if d >= 10]

                # Get corresponding volumes for these indices
                corresponding_volumes = [volumes[i] for i in valid_indices]

                # Create tuples of (volume, index, distance) for sorting
                volume_with_index_and_distance = list(
                    zip(corresponding_volumes, valid_indices, [distances[i] for i in valid_indices])
                )

                # Sort based on volume (largest first) and get the top 10
                top_indices_by_volume = sorted(volume_with_index_and_distance, reverse=True, key=lambda x: x[0])[:10]

                # Store results: index, distance, volume
                results[mutation][f"{run}_{chain}"] = [
                    {"index": idx, "distance": dist, "volume": vol} for vol, idx, dist in top_indices_by_volume
                ]

    return results
```
